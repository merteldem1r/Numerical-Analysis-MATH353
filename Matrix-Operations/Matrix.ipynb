```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
# 2 x 3 matrix
A = np.array([[1, 2, 3],
              [5, 6, 0]])

print("Matrix A:\n", A)
print("Shape of A:", A.shape)
print("Data type of A:", A.dtype)
print("Number of dimensions:", A.ndim)

# 3 x 3 zero matrix
zero3x3 = np.zeros((3, 3))
print("\n3x3 Zero Matrix:\n", zero3x3)

# 3 x 3 ones matrix
ones3x3 = np.ones((3, 3))
print("\n3x3 Ones Matrix:\n", ones3x3)

# 3 x 3 identity matrix
identity3x3 = np.eye(3)
print("\n3x3 Identity Matrix:\n", identity3x3)
```

```python
A = np.array([[1, 2],
              [3, 4]]) # 2x2

B = np.array([[5, 6],
              [7, 8]]) # 2x2

print("Matrix A:\n", A)
print("Matrix B:\n", B)
```

```python
# Matrix Multiplication
M = A * B
print("\nMatrix M (A * B) - Element-wise Multiplication:\n", M)

# This method is not correct for matrix multiplication
# Correct way to do matrix multiplication
M_correct = np.dot(A, B)
print("\nMatrix M_correct (A dot B) - Matrix Multiplication:\n", M_correct)

# or Python method
M_method = A @ B
print("\nMatrix M_method (A @ B) - Matrix Multiplication:\n", M_method)
```

```python
# Matrix Operations

M = np.array([[1, 2], [3, 4]])  # 2x2

print("Matrix M:\n", M)

# 1. Transpose of a matrix
M_transpose = M.T
print("\nTranspose of Matrix M:\n", M_transpose)

# 2. Determinant of a matrix
M_determinant = np.linalg.det(M)
print("\nDeterminant of Matrix A:\n {:.3f}".format(
    M_determinant))  # 1 x 4 - 2 x 3

# 3. Inverse of a matrix
if M_determinant != 0:
    M_inverse = np.linalg.inv(M)
    print("\nInverse of Matrix M:\n", M_inverse)

    # to verify M * M_inv = Identity matrix
    identity_check = np.round(M @ M_inverse)
    print("\nM * M_inv (Should be Identity Matrix):\n", identity_check)
else:
    print("\nMatrix M is singular and cannot be inverted.")
```

```python
# Coefficient Matrix

# 2x2 Coefficient Matrix
A = np.array([[2,1], [3,4]])  # 2x2

# Result Matrix
B = np.array([8, 18])  # 2x1

# 2x + y = 8
# 3x + 4y = 18

# Solution for Ax = B
res = np.linalg.solve(A, B)

print("Solution for Ax = B:\n", res)
print("x = {:.3f}".format(res[0]))
print("y = {:.3f}".format(res[1]))
```

```python
N = 200

print("Generating a random {}x{} matrix...".format(N, N))

A = np.random.rand(N, N)
B = np.random.rand(N, N)

C_loop = np.zeros((N, N))

start_time = time.time()

for i in range(N):  # iterate over rows of A
    for j in range(N):  # iterate over columns of B
        for k in range(N):  # iterate over columns of A / rows of B
            C_loop[i][j] += A[i][k] * B[k][j]

end_time = time.time()
loop_time = end_time - start_time
print("Time taken for matrix multiplication using loops: {:.3f} seconds".format(
    loop_time))

# Using NumPy's built-in function
start_time = time.time()
C_numpy = np.dot(A, B)
end_time = time.time()
numpy_time = end_time - start_time
print("Time taken for matrix multiplication using NumPy: {:.3f} seconds".format(
    numpy_time))

# using @ operator
start_time = time.time()
C_at = A @ B
end_time = time.time()
at_time = end_time - start_time
print("Time taken for matrix multiplication using @ operator: {:.3f} seconds".format(
    at_time))
```

```python
x = np.array([1, 2, 3])
ones = np.ones(len(x))

print("Ones vector:", ones)
print("Vector x:", x)

A = np.vstack((ones, x))
print("A vstack: ", A)

A = A.T
print("A transposed:\n", A)
```

```python
y = np.array([2, 3, 5])

A_T = A.T
A_T_A = A_T @ A
A_T_y = A_T @ y  # result vector

# Solving equation
coeff = np.linalg.solve(A_T_A, A_T_y)
print("Coefficients:\n", coeff)

a = coeff[0]
b = coeff[1]

print("Intercept (a): {:.3f}".format(a))
print("Slope (b): {:.3f}".format(b))
print(f"Equation: y = {b:.4f}x + {a:.4f}")
```

```python
# Generate x points to draw the line
x_line = np.linspace(0, 4, 100)
y_line = a + b * x_line
plt.figure(figsize=(8, 5))
# Plot the original points
plt.scatter(x, y, color='red', s=100, label='Given Points (1,2), (2,3), (3,5)')
# Plot the best fit line we found
plt.plot(x_line, y_line, color='blue', linewidth=2, label=f'Least Squares Line\ny = {b:.2f}x + {a:.2f}')
plt.title('Line Fitting with Least Squares Method')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
